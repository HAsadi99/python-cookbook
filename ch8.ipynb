```python
# 8.2 Saving Images

# oad image as grayscale
image = cv2.imread('images.jpeg', cv2.IMREAD_GRAYSCALE)

# save image
cv2.imwrite('images.jpeg' , image)
```

```python
# 8.3 Resizing Images

# load image as grayscale
image = cv2.imread('images.jpeg', cv2.IMREAD_GRAYSCALE)

# Resize image to 50 pixels by 50 pixels
image_70_70 = cv2.resize(image , (70, 70))

# view image
plt.imshow(image_70_70 , cmap='gray')
plt.axis('off')
plt.show()
```

```python
# 8.4 Cropping Images

image = cv2.imread('images.jpeg' , cv2.IMREAD_GRAYSCALE)

# crop image with select
image_cropped = image[: , :180]

# show image
plt.imshow(image_cropped, cmap = "gray")
plt.axis('off')
plt.show()
```

```python
# 8.5 Blurring Images

# load image in gray style
image = cv2.imread('images.jpeg' , cv2.IMREAD_GRAYSCALE)

# blur image
image_blurry = cv2.blur(image, (5,5))

# show image
plt.imshow(image_blurry, cmap = "gray")
plt.axis('off')
```

```python
image_very_blurry = cv2.blur(image, (50,50))

# show image
plt.imshow(image_very_blurry, cmap = "gray")
plt.axis('off')
```

```python
# create kernel
kernel = np.ones((5,5)) / 25

kernel
```

```python
# Apply kernel
image_kernel = cv2.filter2D(image, -1, kernel)

plt.imshow(image_kernel, cmap = "gray")
plt.axis('off')
plt.xticks([])
```

```python
# 8.6 Sharpening Images
# load image in gray style
image = cv2.imread('images.jpeg' , cv2.IMREAD_GRAYSCALE)

# create kernel
kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])

# sharpen image
image_sharp = cv2.filter2D(image, -1, kernel)

# show image
plt.imshow(image_sharp, cmap = "gray")
plt.axis('off')
```
